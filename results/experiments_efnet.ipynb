```python
import pickle
import matplotlib.pyplot as plt
```

```python
with open('EfNetB0_2.1_losses.pkl', 'rb') as f:
    losses_efnet = pickle.load(f)

with open('EfNetB0_2.1_metrics.pkl', 'rb') as f:
    metrics_efnet = pickle.load(f)
```

```python
fig, axes = plt.subplots(2, 1, figsize=(12, 12))
fig.suptitle('EfficienNetB2', fontsize='20')

axes[0].plot(losses_efnet['val'], label='val')
axes[0].plot(losses_efnet['train'], label='train')
axes[1].plot(metrics_efnet['val'], label='val')
axes[1].plot(metrics_efnet['train'], label='train')
axes[0].legend()
axes[1].legend()
axes[0].set_xlabel('Num of epoch')
axes[0].set_ylabel('Val Loss')
axes[1].set_xlabel('Num of epoch')
axes[1].set_ylabel('Val accuracy')
```
