```python
import pickle
import matplotlib.pyplot as plt
```

```python
with open('ResNet_exp/resnet_50_20ep_losses.pkl', 'rb') as f:
    losses_res_50_20 = pickle.load(f)

with open('ResNet_exp/resnet_50_20ep_metrics.pkl', 'rb') as f:
    metrics_res_50_20 = pickle.load(f)

with open('ResNet_exp/resnet_losses.pkl', 'rb') as f:
    losses_res_18 = pickle.load(f)

with open('ResNet_exp/resnet_metrics.pkl', 'rb') as f:
    metrics_res_18 = pickle.load(f)
    
with open('ResNet_exp/resnet_50 (without)_losses.pkl', 'rb') as f:
    losses_res_without = pickle.load(f)
    
with open('ResNet_exp/resnet_50 (without)_metrics.pkl', 'rb') as f:
    metrics_res_without = pickle.load(f) 
    
with open('ResNet_exp/resnet_50_SE_metrics.pkl', 'rb') as f:
    resnet_se_metrics = pickle.load(f) 

with open('ResNet_exp/resnet_50_SE_losses.pkl', 'rb') as f:
    resnet_se_losses = pickle.load(f) 

with open('ResNet_exp/resnet_50_ls_losses.pkl', 'rb') as f:
    resnet_ls_losses = pickle.load(f) 

with open('ResNet_exp/resnet_50_ls_metrics.pkl', 'rb') as f:
    resnet_ls_metrics = pickle.load(f) 
```

### ResNets 

1) resnet18: \
num of epochs: 10
2) resnet 50: \
num of epochs: 20

For each model except for resnet with se and resnet with ls the batch_size = 16


```python
fig, axes = plt.subplots(4, 1, figsize=(12, 12))
fig.suptitle('Different versions of ResNet', fontsize='20')

axes[0].plot(losses_res_50_20['train'], label='resnet50 with trick')
axes[0].plot(losses_res_18['train'], label='resnet18')
axes[0].plot(losses_res_without['train'], label='resnet50 without trick')
axes[0].plot(resnet_se_losses['train'], label='resnet50 with se')
axes[0].plot(resnet_ls_losses['train'], label='resnet50 with ls')
axes[0].legend()
axes[0].set_xlabel('Num of epoch')
axes[0].set_ylabel('Train Loss')

axes[1].plot(losses_res_50_20['val'], label='resnet50 with trick')
axes[1].plot(losses_res_18['val'], label='resnet18')
axes[1].plot(losses_res_without['val'], label='resnet50 without trick')
axes[1].plot(resnet_se_losses['val'], label='resnet50 with se')
axes[1].plot(resnet_ls_losses['val'], label='resnet50 with ls')
axes[1].legend()
axes[1].set_xlabel('Num of epoch')
axes[1].set_ylabel('Val Loss')

axes[2].plot(metrics_res_50_20['train'], label='resnet50 with trick')
axes[2].plot(metrics_res_18['train'], label='resnet18')
axes[2].plot(metrics_res_without['train'], label='resnet50 without trick')
axes[2].plot(resnet_se_metrics['train'], label='resnet50 with se')
axes[2].plot(resnet_ls_metrics['train'], label='resnet50 with ls')
axes[2].legend()
axes[2].set_xlabel('Num of epoch')
axes[2].set_ylabel('Train Accuracy')

axes[3].plot(metrics_res_50_20['val'], label='resnet50 with trick')
axes[3].plot(metrics_res_18['val'], label='resnet18')
axes[3].plot(metrics_res_without['val'], label='resnet50 without trick')
axes[3].plot(resnet_se_metrics['val'], label='resnet50 with se')
axes[3].plot(resnet_ls_metrics['val'], label='resnet50 with ls')
axes[3].legend()
axes[3].set_xlabel('Num of epoch')
axes[3].set_ylabel('Val Accuracy')

plt.show()
```
